# Advanced NLP Tokenization Demo

This notebook demonstrates sentence and word tokenization using a modular, class-based approach with NLTK.


```python
# Dependency Setup

import subprocess
import sys

def install_package(pkg_name):
    try:
        __import__(pkg_name)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg_name])

install_package("nltk")
```

```python
# Imports & Downloads

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

def initialize_nltk():
    required_resources = ["punkt"]
    for resource in required_resources:
        try:
            nltk.data.find(f"tokenizers/{resource}")
        except LookupError:
            nltk.download(resource)

initialize_nltk()
```

```python
#Sample Text

def get_sample_text():
    return (
        "World War III, also known as the Third World War, is a hypothetical future "
        "global conflict subsequent to World War I and World War II. It is widely "
        "predicted that such a war would involve advanced weaponry and widespread destruction."
    )

text_data = get_sample_text()
```

```python
#  NLP Processing Class

class TextProcessor:
    
    def __init__(self, text: str):
        self.text = text
    
    def sentence_segmentation(self):
        """Break text into sentences."""
        return sent_tokenize(self.text)
    
    def word_tokenization(self):
        """Break text into words."""
        return word_tokenize(self.text)
    
    def process_all(self):
        """Run all NLP steps."""
        return {
            "sentences": self.sentence_segmentation(),
            "words": self.word_tokenization()
        }
```

```python
#  Execution

if __name__ == "__main__":
    processor = TextProcessor(text_data)
    results = processor.process_all()
    
    print("\n📌 Sentences:\n", results["sentences"])
    print("\n📌 Words:\n", results["words"])
```
