```python
# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read Image
img = cv2.imread("Sample.jpg")
```

```python
# Convert BGR to RGB 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Apply Canny Edge Detection
edges = cv2.Canny(gray, 100, 200)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edges")
plt.axis("off")
plt.show()
```

```python
resized = cv2.resize(img_rgb, (300, 300))

plt.imshow(resized)
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img_rgb, matrix, (w, h))

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
cropped = img_rgb[50:300, 50:300]

plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis("off")
plt.show()

```

```python
shape_img = img_rgb.copy()

cv2.rectangle(shape_img, (50,50), (200,200), (255,0,0), 3)
cv2.circle(shape_img, (300,150), 50, (0,255,0), 3)
cv2.line(shape_img, (0,0), (400,400), (0,0,255), 3)

plt.imshow(shape_img)
plt.title("Shapes")
plt.axis("off")
plt.show()
```

```python
text_img = img_rgb.copy()

cv2.putText(text_img, "Talha Usman ", (50,50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (225,0,0), 2)

plt.imshow(text_img)
plt.title("Text on Image")
plt.axis("off")
plt.show()
```

```python

blur = cv2.GaussianBlur(img_rgb, (15,15), 0)

plt.imshow(blur)
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Binary Threshold")
plt.axis("off")
plt.show()
```

```python
rows, cols = img.shape[:2]

matrix = np.float32([[1, 0, 100], [0, 1, 50]])
translated = cv2.warpAffine(img_rgb, matrix, (cols, rows))

plt.imshow(translated)
plt.title("Translated Image")
plt.axis("off")
plt.show()
```

```python
flip_horizontal = cv2.flip(img_rgb, 1)

plt.imshow(flip_horizontal)
plt.title("Flipped Image")
plt.axis("off")
plt.show()
```

```python
plt.hist(gray.ravel(), 256, [0,256])
plt.title("Image Histogram")
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray, 1.3, 5)

face_img = img_rgb.copy()

for (x,y,w,h) in faces:
    cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,0,0), 2)

plt.imshow(face_img)
plt.title("Face Detection")
plt.axis("off")
plt.show()
```

```python
contours, hierarchy = cv2.findContours(
    thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

contour_img = img_rgb.copy()
cv2.drawContours(contour_img, contours, -1, (0,255,0), 2)

plt.imshow(contour_img)
plt.title("Contours")
plt.axis("off")
plt.show()
```
