```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text = """World War III,[a] also known as the Third World War, is a hypothetical future global conflict subsequent to World War I (1914–1918) and World War II (1939–1945). It is widely predicted that such a war would involve all of the great powers, like its two predecessors, and the use of nuclear weapons or other weapons of mass destruction, thereby surpassing all prior conflicts in scale, devastation, and loss of life.
World War III was initially synonymous with the escalation of the Cold War (1947–1991) into conflict between the US-led Western Bloc and Soviet-led Eastern Bloc, though the United States and the Soviet Union never directly engaged. Since the US's development and use of nuclear weapons in the atomic bombings of Hiroshima and Nagasaki at the end of World War II, the risk of a nuclear apocalypse causing widespread destruction and the potential collapse of modern civilization or human extinction has been central in speculation and fiction about World War III. The Soviet Union's development of nuclear weapons in 1949 spurred the nuclear arms race and was followed by several other countries.
"""
text
```

```python
# 1. Segmentation : Split text into sentences
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(text)
sentences
```

```python
# 2. Tokenization : Split sentences into words
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
words
```

```python
import re
text2 = re.sub(r"[^a-zA-Z0-9]", " ", text)
print(text2)
```

```python
# 3. Stop Words : is, am, are, was, were, he, she, it, all, about, above, everything
nltk.download('stopwords')
from nltk.corpus import stopwords

# print(stopwords.words('english'))

# clean_words = []
# for i in words:
#   if i not in stopwords.words('english'):
#     clean_words.append(i)

clean_words = [i for i in words if i not in stopwords.words('english')]

# clean_words = [i for i in words if i not in stopwords.words('english')]
# clean_words





# for w in words:
#   if w not in stopwords.words('english') or w in ['in', 'my', 'her', 'his', 'their']:
#     words.append(w)

# words = [w for w in words if w not in stopwords.words('english')]
# print(words)
# print(text)
```

```python
# 4. Stemming / Lemmatization
# 4.1. Stemming      -> Stemma : "combining" -> "combin", "cried" = "cri"
from nltk.stem.porter import PorterStemmer

# stemmed = []
# for w in words:
#   stemmed.append(PorterStemmer().stem(w))

stemmed     = [PorterStemmer().stem(w) for w in words]
print(stemmed)
```

```python
# 4.2. Lemmatization -> Lemma  : "combining" -> "combine", "cried" = "cry"
nltk.download('wordnet')
# nltk.download('omw-1.4')

from nltk.stem.wordnet import WordNetLemmatizer
lemmatized = [WordNetLemmatizer().lemmatize(w) for w in words]
print(lemmatized)
```

```python
# 5. Part of Speech (POS) Tagging :
nltk.download('averaged_perceptron_tagger_eng')

from nltk import pos_tag
tags = pos_tag(lemmatized)
tags
```

```python
"""
POS
CC: It is the conjunction of coordinating
CD: It is a digit of cardinal
DT: It is the determiner
EX: Existential
FW: It is a foreign word
IN: Preposition and conjunction
JJ: Adjective
JJR and JJS: Adjective and superlative
LS: List marker
MD: Modal
NN: Singular noun
NNS, NNP, NNPS: Proper and plural noun
PDT: Predeterminer
WRB: Adverb of wh
WP$: Possessive wh
WP: Pronoun of wh
WDT: Determiner of wp
VBZ: Verb
VBP, VBN, VBG, VBD, VB: Forms of verbs
UH: Interjection
TO: To go
RP: Particle
RBS, RB, RBR: Adverb
PRP, PRP$: Pronoun personal and professional
"""
```

```python
# 6. Named Entity Recognition (NER) :
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')

from nltk import ne_chunk

ner = ne_chunk(tags)
print(ner)
```

```python
def analyze_sentiment(text):
  sentiment_score = sia.polarity_scores(text)
  if sentiment_score['compound'] >= 0.05:
    return "Positive Sentence"
  elif sentiment_score['compound'] <= -0.05:
    return "Negative Sentence"
  else:
    return "Neutral Sentence"
```

```python
def chatbot_app():
  print(" -- Welcome to BSAI-4B Chatbot -- ")
  print(" -- Type 'sentiment' for Sentiment Analysis")
  print(" -- Type 'Exit' to exit")

  while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
      print("ChatBot: Goodbye!")
      break
    elif user_input.lower() == "sentiment":
      print("ChatBot: Tell me a sentence to Analyze.")
      sentiment_input = input("You: ")
      print("ChatBot: Sentiment Analysis Report -", analyze_sentiment(sentiment_input))
    else:
      response = chatbot.respond(user_input.lower())
      if response:
        print("ChatBot:", response)
        if "bye" in user_input.lower():
          break
      else:
        print("ChatBot: I'm not sure how to respond to that.")
```
