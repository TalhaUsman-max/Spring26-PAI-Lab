```python
import pandas as pd
import numpy as np


class AuthenticationDataManager:
    """
    Handles authentication table transformations
    and password integration operations.
    """

    def __init__(self, account_frame: pd.DataFrame):
        self.account_frame = account_frame

    # ---------------------------------------------------
    # Remove Deprecated Verification Column
    # ---------------------------------------------------

    def discard_verification_status(self) -> None:
        """
        Removes the old verification field
        from the authentication dataset.
        """

        removable_column = "Verified"

        if removable_column in self.account_frame.columns:
            self.account_frame.drop(
                columns=[removable_column],
                inplace=True
            )

    # ---------------------------------------------------
    # Inject Password Hashes
    # ---------------------------------------------------

    def attach_password_hashes(
        self,
        credential_matrix: np.ndarray
    ) -> None:
        """
        Adds password hashes extracted
        from the NumPy credential matrix.
        """

        extracted_hashes = credential_matrix[:, 1]

        self.account_frame.insert(
            loc=len(self.account_frame.columns),
            column="EncryptedPassword",
            value=extracted_hashes
        )

    # ---------------------------------------------------
    # Full Transformation Pipeline
    # ---------------------------------------------------

    def build_secure_login_table(
        self,
        credential_matrix: np.ndarray
    ) -> pd.DataFrame:
        """
        Executes the complete authentication
        table enhancement workflow.
        """

        self.discard_verification_status()
        self.attach_password_hashes(credential_matrix)

        return self.account_frame


# =======================================================
# Sample Authentication Dataset
# =======================================================

user_authentication_records = pd.DataFrame({
    "UserID": [101, 102],
    "Username": ["sara", "talha"],
    "Verified": [True, False]
})


# =======================================================
# Password Hash Storage Matrix
# =======================================================

hashed_credentials_array = np.array([
    [101, "secure_hash_alpha"],
    [102, "secure_hash_beta"]
])


# =======================================================
# Execute Authentication Transformation
# =======================================================

security_pipeline = AuthenticationDataManager(
    user_authentication_records
)

updated_login_dataset = (
    security_pipeline.build_secure_login_table(
        hashed_credentials_array
    )
)

print(updated_login_dataset)
```
