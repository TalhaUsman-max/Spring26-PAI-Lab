```python
# Import important libraries
import pandas as pd
import re

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Create simple sentiment dataset
reviews_data = {
    "reviews": [
        "I love this Pakistan",
        "This is the my favourite place",
        "Amazing experience",
        "My name is Talha Usman and I am a student of computer science",
        "I am a truck dispatcher and I am loving it",
        "Not worth the money",
        "Absolutely fantastic",
        "Terrible service"
    ],

    # 1 = Positive , 0 = Negative
    "sentiment": [1, 0, 1, 0, 1, 0, 1, 0]
}

# Convert dictionary into dataframe
dataset = pd.DataFrame(reviews_data)

# Function for cleaning text
def text_cleaner(sentence):

    # Convert into lowercase
    sentence = sentence.lower()

    # Remove numbers and symbols
    sentence = re.sub(r'[^a-z]', ' ', sentence)

    return sentence

# Apply cleaning function
dataset["clean_reviews"] = dataset["reviews"].apply(text_cleaner)

# Input and Output data
x = dataset["clean_reviews"]
y = dataset["sentiment"]

# Split data into training and testing
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.25,
    random_state=1
)

# Convert text into numbers
text_converter = CountVectorizer()

x_train_features = text_converter.fit_transform(x_train)
x_test_features = text_converter.transform(x_test)

# Create model
sentiment_model = LogisticRegression()

# Train model
sentiment_model.fit(x_train_features, y_train)

# Make predictions
predicted_output = sentiment_model.predict(x_test_features)

# Check accuracy
score = accuracy_score(y_test, predicted_output)

print("Model Accuracy:", score)

# Detailed report
print("\nClassification Report:\n")
print(classification_report(y_test, predicted_output))

# Test model with new sentence
my_review = ["This product is really amazing and good"]

# Convert into vector
my_review_vector = text_converter.transform(my_review)

# Predict sentiment
final_result = sentiment_model.predict(my_review_vector)

# Show result
if final_result[0] == 1:
    print("\nPrediction: Positive Review")
else:
    print("\nPrediction: Negative Review")
```
