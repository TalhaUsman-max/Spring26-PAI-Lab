```python
# Water Jug Problem using DFS 

jug1_capacity = int(input("Enter Jug1 capacity: "))
jug2_capacity = int(input("Enter Jug2 capacity: "))
goal = int(input("Enter Goal amount: "))

visited = set()

def dfs(jug1, jug2, path):

    if (jug1, jug2) in visited:
        return False

    visited.add((jug1, jug2))

    # Goal condition
    if jug1 == goal or jug2 == goal:
        print("\nSolution Found\nSteps:")
        for i, step in enumerate(path, 1):
            print(f"Step {i}: {step}")
        print("Final State:", (jug1, jug2))
        print("Total Steps:", len(path))
        return True

    if dfs(jug1_capacity, jug2, path + [f"Fill Jug1 -> ({jug1_capacity},{jug2})"]):
        return True

    if dfs(jug1, jug2_capacity, path + [f"Fill Jug2 -> ({jug1},{jug2_capacity})"]):
        return True

    if dfs(0, jug2, path + [f"Empty Jug1 -> (0,{jug2})"]):
        return True

    if dfs(jug1, 0, path + [f"Empty Jug2 -> ({jug1},0)"]):
        return True

    transfer = min(jug1, jug2_capacity - jug2)
    if dfs(jug1 - transfer, jug2 + transfer,
           path + [f"Pour Jug1 -> Jug2 -> ({jug1-transfer},{jug2+transfer})"]):
        return True

    transfer = min(jug2, jug1_capacity - jug1)
    if dfs(jug1 + transfer, jug2 - transfer,
           path + [f"Pour Jug2 -> Jug1 -> ({jug1+transfer},{jug2-transfer})"]):
        return True

    return False

# Start from empty jugs
dfs(0, 0, ["Start -> (0,0)"])
```
