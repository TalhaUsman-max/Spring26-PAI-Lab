###  Simple Water Jug Problem with rule choice 


```python
x = int(input("Enter capacity of Jug X: "))
y = int(input("Enter capacity of Jug Y: "))
goal = int(input("Enter goal: "))

X = 0
Y = 0

step = 0

while True:
    print("\nChoose Rule:")
    print("1. Fill X")
    print("2. Fill Y")
    print("3. Empty X")
    print("4. Empty Y")
    print("5. Pour X into Y until X is empty")
    print("6. Pour X into Y until Y is full")
    print("7. Pour Y into X until Y is empty")
    print("8. Pour Y into X until X is full")
    print("0. Exit")

    choice = int(input("Enter choice: "))

    if choice == 0:
        print("Exited")
        break

    step += 1

    if choice == 1:
        X = x

    elif choice == 2:
        Y = y

    elif choice == 3:
        X = 0

    elif choice == 4:
        Y = 0

    elif choice == 5:
        transfer = min(X, y - Y)
        X -= transfer
        Y += transfer

    elif choice == 6:
        transfer = min(X, y - Y)
        X -= transfer
        Y += transfer

    elif choice == 7:
        transfer = min(Y, x - X)
        Y -= transfer
        X += transfer

    elif choice == 8:
        transfer = min(Y, x - X)
        Y -= transfer
        X += transfer

    print(f"Step {step}: ({X}, {Y})")

    if X == goal or Y == goal:
        print("Goal Reached!")
        break
```

###  Simple Water Jug Problem with help of DFS with rules

```python
x = int(input("Enter Jug X capacity: "))
y = int(input("Enter Jug Y capacity: "))
goal = int(input("Enter Goal: "))

stack = [((0, 0), [])]
visited = set()

while stack:
    (X, Y), path = stack.pop()

    if (X, Y) in visited:
        continue

    visited.add((X, Y))

    if X == goal or Y == goal:
        path.append((X, Y, "Goal Reached"))
        for i, (sx, sy, rule) in enumerate(path, 1):
            print(f"Step {i}: {rule} -> ({sx}, {sy})")
        break

    moves = [
        ("Fill X", (x, Y)),
        ("Pour X -> Y", (X - min(X, y - Y), Y + min(X, y - Y))),
        ("Empty Y", (X, 0)),
        ("Pour Y -> X", (X + min(Y, x - X), Y - min(Y, x - X))),
        ("Empty X", (0, Y)),
        ("Fill Y", (X, y)),
    ]

    for rule, (nx, ny) in moves:
        if (nx, ny) not in visited:
            new_path = path.copy()
            new_path.append((nx, ny, rule))   # <-- FIX HERE
            stack.append(((nx, ny), new_path))
```
